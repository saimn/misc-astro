```python
def expplt(ax, cmap=None):
    x, y = np.mgrid[0:512,0:512]
    return ax.imshow(np.exp( -((x-256)**2 + (y-256)**2)/1e6), cmap=cmap)
    #plt.colorbar()
```

```python
def sinplot(ax, flip=1):
    x = np.linspace(0, 14, 100)
    for i in range(1, 7):
        ax.plot(x, np.sin(x + i * .5) * (7 - i) * flip)
```

```python
fig, ax = plt.subplots()
axim = expplt(ax, cmap='viridis')
fig.colorbar(axim, ax=ax)
```

```python
for i, style in enumerate(sorted(plt.style.available)):
    with plt.style.context(style):
        fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(12,5))
        sinplot(ax1)
        ax1.set_title(style)
        expplt(ax2)
        plt.show()
```
