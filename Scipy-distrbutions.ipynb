# Scipy distributions

https://docs.scipy.org/doc/scipy/reference/generated/scipy.stats.norm.html#scipy.stats.norm

```python
%matplotlib inline

import matplotlib.pyplot as plt
import numpy as np
from scipy.stats import norm
```

```python
norm.ppf([0.001, 0.5, 0.999])
```

```python
norm.cdf(norm.ppf([0.001, 0.5, 0.999]))
```

```python
fig, (ax1, ax2, ax3) = plt.subplots(1, 3, figsize=(16, 4))

x = np.linspace(0, 1, 101)
ax1.plot(x, norm.ppf(x))
ax1.set(title='ppf - Percent point function\n (inverse of cdf â€” percentiles)')

x = np.linspace(-3, 3, 101)
ax2.plot(x, norm.cdf(x))
ax2.set(title='cdf - Cumulative distribution function')

x = np.linspace(norm.ppf(0.01), norm.ppf(0.99), 100)
ax3.plot(x, norm.pdf(x), 'r-', lw=5, alpha=0.5, label='norm pdf')

rv = norm()
ax3.plot(x, rv.pdf(x), 'k-', lw=2, label='frozen pdf')

# Generate random numbers:
r = norm.rvs(size=1000)
ax3.hist(r, density=True, bins=30, histtype='stepfilled', alpha=0.5)

ax3.set(title='pdf')
ax3.legend(loc='best', frameon=False);
```
